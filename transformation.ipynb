```python
players = pd.read_csv('data/perfectgame/perfectgame_top50_2002-2021_raw.csv')

to_st = {}
for i in players.college.unique():
    if 'State' in i:
        to_st[i] = i.replace('State', 'St.') 

hardcoded_map = {'Louisiana St.':'LSU',
                 'Georgia Southern':'Ga. Southern',
                 'Miami':'Miami (FL)', 
                 'Florida International':'FIU',
                 'Mississippi':'Ole Miss',
                 'North Carolina St.':'NC State',
                 'Southern':'Southern U.',
                 'Southern Miss':'Southern Miss.',
                 'Southeastern Louisiana':'Southeastern La.',
                 'College of Charleston':'Col. of Charleston',
                 'Florida Atlantic':'Fla. Atlantic',
                 'University of Connecticut':'UConn',
                 'South Florida':'South Fla.',
                 'Dallas Baptist':'DBU',
                 'Loyola Marymount':'LMU (CA)',
                 'Brigham Young':'BYU',
                 'Illinois-Chicago':'UIC',
                 "St. John's":"St. John's (NY)",
                 'UNC-Wilmington':'UNCW'
                 }
                 # > 'Louisiana-Lafayette':}

players.college = players.college.map(to_st).fillna(players.college)
players.college = players.college.map(hardcoded_map).fillna(players.college)
master = pd.read_csv('data/ncaa/school_lookup.csv')
schools = master[['school', 'school_id']]
unique_schools = schools.drop_duplicates()
all_players = players.merge(unique_schools, how='left', left_on='college', right_on='school')

with_college = all_players[all_players.college != 'None']
no_college = all_players[all_players.college == 'None']
mapped_to_school = with_college[with_college.school_id.notna()]
not_mapped = with_college[with_college.school_id.isna()]
```
